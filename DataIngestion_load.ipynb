```python
from langchain_community.document_loaders import TextLoader
loadder = TextLoader('/content/speech.txt')
text_documents=loadder.load()

for i in text_documents:
    print(i)
```

```python
# split the textual data in chunks for LLM model
from langchain_text_splitters import CharacterTextSplitter
text_splitter = CharacterTextSplitter(separator="\n\n",chunk_size=20,chunk_overlap=20)
text=text_splitter.split_documents(text_documents)
text
```
